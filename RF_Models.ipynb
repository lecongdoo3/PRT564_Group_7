```python
import matplotlib.pyplot as plt

# Data for the models and their accuracy
models = ['Baseline model (NB)', 'RF default', 'RF after random undersampling', 'RF after SMOTE data balancing', 'RF after optimisation']
accuracy = [46.95, 74.85, 73.03, 78.41, 79.00]

# Creating the line graph with updates
plt.figure(figsize=(10, 5))
plt.plot(models, accuracy, marker='o', color='orange', linewidth=2)  # Setting the line color to yellow
plt.title('Changes in Performance of Random Forest after Applying Different Techniques')
plt.ylabel('Accuracy (%)')
plt.ylim(40, 85)
plt.grid(True)
plt.xticks(rotation=45)

# Adding data labels to each point
for i, acc in enumerate(accuracy):
    plt.text(models[i], acc + 1, f'{acc}%', ha='center')

plt.show()
```
