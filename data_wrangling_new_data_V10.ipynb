#### Remove unnecessary columns

```python
columns_to_remove = ['URLS', 'RetractionDOI', 'RetractionPubMedID', 'OriginalPaperDOI', 
                     'OriginalPaperPubMedID', 'RetractionNature', 'Notes']

# Remove the specified columns from the DataFrame
df = df.drop(columns=columns_to_remove, errors='ignore')
```

#### Fill Missing Values

```python
# Fill missing 'Institution' values with "Unknown"
df['Institution'] = df['Institution'].fillna('Unknown')

# Because 'Paywalled' is binary data, fill missing values with the most common value (mode)
# Check mode
most_common_paywalled = df['Paywalled'].mode()[0]

# Then, fill missing 'Paywalled' values with the most common value
df['Paywalled'] = df['Paywalled'].fillna(most_common_paywalled)
```

```python
# Recheck null values to see if they're all filled.
df.isnull().sum()
```

Now there's no missing data in the dataset.

#### Check date format if it's consistant

```python
# Removing " 0:00" from specified columns
df['RetractionDate'] = df['RetractionDate'].str.replace(' 0:00', '', regex=False)
df['OriginalPaperDate'] = df['OriginalPaperDate'].str.replace(' 0:00', '', regex=False)
```

```python
# Convert and format the dates in the columns
columns_to_update = ['RetractionDate', 'OriginalPaperDate']
for column in columns_to_update:
    df[column] = pd.to_datetime(df[column], format='%m/%d/%Y').dt.strftime('%d/%m/%Y')
```

#### Check unknown values

```python
specified_list = ['unavailable', 'unknown']

# Function to process each value
def process_value(x):
    return x.lower() in specified_list if isinstance(x, str) else False

# Applying the function to each element in the DataFrame while preserving column names
processed_df = df.copy()
for col in processed_df.columns:
    processed_df[col] = processed_df[col].map(process_value)

# Counting True values for each column
counts = processed_df.sum()

# To get the total count across the whole DataFrame, sum up the column counts
total_count = counts.sum()

print("Counts per column:")
print(counts)
print("\nTotal count across the DataFrame:", total_count)
```

```python
# Function to replace specified values with 'Unknown'
def replace_with_unknown(x):
    return 'Unknown' if isinstance(x, str) and x.lower() in specified_list else x

# Applying the function to each element in the DataFrame
for col in df.columns:
    df[col] = df[col].map(replace_with_unknown)
```

```python
# Counting 'Unknown' values for each column
counts = (df == 'Unknown').sum()

# To get the total count across the whole DataFrame, sum up the column counts
total_count = counts.sum()

counts, total_count
```

#### 5. Reason Group for Retraction


This method is based on Scientific Error and Academic Misconduct.

Group 1: Scientific error and academic misconduct 

Group 2: Academic misconduct and no scientific error

Group 3: Scientific error and no academic misconduct

Group 4: No scientific error and no academic misconduct

```python
# Define keywords dictionary for each group
group1_keywords = [
    'Falsification/Fabrication of Data', 'Falsification/Fabrication of Image', 'Falsification/Fabrication of Results', 
    'Manipulation of Images', 'Manipulation of Results', 'Plagiarism of Data', 'Plagiarism of Image', 'Fake Peer Review',
    'Sabotage of Materials'
]

group2_keywords = [
    'Duplication of Article', 'Duplication of Data', 'Duplication of Image', 'Duplication of Text', 
    'Euphemisms for Duplication', 'Euphemisms for Misconduct', 'Euphemisms for Plagiarism', 'Forged Authorship',
    'Misconduct – Official Investigation/Finding', 'Plagiarism of Article', 'Plagiarism of Data', 'Plagiarism of Image', 
    'Plagiarism of Text', 'Rogue Editor', 'Salami Slicing', 'Taken from Dissertation/Thesis'
]

group3_keywords = [
    'Concerns/Issues About Data', 'Concerns/Issues About Image', 'Concerns/Issues About Results ', 'Contamination of Cell Lines/Tissues',
    'Contamination of Materials (General)', 'Contamination of Reagents', 'Error in Analyzes', 'Error in Cell Lines/Tissues', 
    'Error in Data', 'Error in Image', 'Error in Materials (General)', 'Error in Methods', 'Error in Results and/or Conclusions', 
    'Results Not Reproducible', 'Unreliable Data', 'Unreliable Image', 'Unreliable Results'
]

group4_keywords = [
    'Author Unresponsive', 'Breach of Policy by Author', 'Breach of Policy by Third Party', 'Complaints about Author', 
    'Complaints about Company/Institution', 'Complaints about Third Party', 'Conflict of Interest', 'Error by Journal/Publisher', 
    'Error by Third Party', 'Error in Text', 'Investigation by Company/Institution', 'Investigation by Journal/Publisher', 
    'Investigation by ORI', 'Investigation by Third Party', 'Lack of Approval from Author', 'Lack of Approval from Company /Institution', 
    'Lack of Approval from Third Party', 'Lack of IRB/IACUC Approval', 'Notice – Lack of', 'Notice – Limited or No Information', 
    'Notice – Unable to Access via current resources', 'Objections by Author(s)', 'Objections by Company/Institution', 
    'Objections by Third Party', 'Original Data not Provided', 'Paper Mill', 'Publishing Ban', 'Results Not Reproducible', 
    'Retract and Replace', 'Temporary Removal', 'Transfer of Copyright/Ownership', 'Updated to Correction', 'Updated to Retraction', 
    'Upgrade/ Update of Prior Notice', 'Withdrawal', 'Withdrawn (out of date)', 'Withdrawn to Publish in Different Journal'
]

group1 = "Scientific error and academic misconduct"
group2 = "Academic misconduct and no scientific error"
group3 = "Scientific error and no academic misconduct"
group4 = "No scientific error and no academic misconduct"

# Function to assign each row of reason column with its group number
def get_groups(reason_string):
    groups = []
    if any(keyword in reason_string for keyword in group1_keywords):
        groups.append(group1)  # Assigning to Group 1
    if any(keyword in reason_string for keyword in group2_keywords):
        groups.append(group2)  # Assigning to Group 2
    if any(keyword in reason_string for keyword in group3_keywords):
        groups.append(group3)  # Assigning to Group 3
    if (any(keyword in reason_string for keyword in group4_keywords)) or (not groups):
        groups.append(group4)  # Assigning to Group 4

    return ';'.join(groups)  # Return a string with groups separated by semicolons

# Applying the function to each row in the DataFrame
df['Reason_group'] = df['Reason'].apply(get_groups)
```

#### 10. LevelOfSeriousness

```python
# Function to determine the seriousness of each retraction
def determine_seriousness(row):
    # Print the reason group for the current row for debugging
    # print(f"Current Reason Group: {row['Reason_group']}")
    
    # Split the reason group string by semicolon
    reason_groups = row['Reason_group'].split(';')
    
    # Determine the level of seriousness based on the reason groups
    if group1 in reason_groups and len(reason_groups) <= 4:
        print('The most serious')
        return 'The most serious'
    if group2 in reason_groups and group3 in reason_groups and len(reason_groups) < 4:
        print('The most serious')
        return 'The most serious'
    if group2 in reason_groups and len(reason_groups) == 1:
        print('Serious')
        return 'Serious'
    if group2 in reason_groups and group4 in reason_groups and len(reason_groups) == 2:
        print('Serious')
        return 'Serious'
    if group3 in reason_groups and group4 in reason_groups and len(reason_groups) == 2:
        print('Less serious')
        return 'Less serious'
    if group3 in reason_groups and len(reason_groups) == 1:
        print('Less serious')
        return 'Less serious'
    if group4 in reason_groups and len(reason_groups) == 1:
        print('The least serious')
        return 'The least serious'
    else:
        print('Unknown')
        return 'Seriousness unknown'  # Handle other cases or when no group is mentioned

# Apply the function to each row and store the result in a new column
df['LevelOfSeriousness'] = df.apply(determine_seriousness, axis=1)
```

```python
# Count values in the columns "LevelOfSeriousness"
seriousness_counts = df['LevelOfSeriousness'].value_counts()

# Check if there is still "Seriousness unknown"
print("Count of 'Seriousness unknown':", seriousness_counts.get("Seriousness unknown", 0))
```
